```python
import dhlab as dh
import pandas as pd
```

```python
with open("markens_grøde.txt") as fp:
    txt = fp.read()
```

```python
toks = txt.split()
```

```python
pairs = []
for i,w in enumerate(toks):
    pairs.append((i,w))
```

```python
work = pd.DataFrame(pairs, columns=["pos","token"])
```

```python
df = pd.DataFrame()
for i in range(0,len(work), 1000):
    d = work[work.index < i]
    g = d.groupby("token")
    c = g.sum()/g.count()
    c.columns = [f"c{i}"]
    df = pd.concat([df, c], axis=1)
```

```python
df = pd.DataFrame()
for i in range(0, len(work), 1000):
    # Take a slice from i to i+1000
    d = work.iloc[i:i+1000]
    g = d.groupby("token")
    c = g.sum() / g.count()
    c.columns = [f"c{i}"]
    df = pd.concat([df, c], axis=1)
```
